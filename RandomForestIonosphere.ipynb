```python
from sklearn.ensemble import RandomForestClassifier
def n_estim(tst_siz,n_value):
    X_train, X_test, y_train, y_test = train_test_split(x_val,y_val, test_size=tst_siz,random_state=101)
    algo=RandomForestClassifier(n_estimators=n_value)
    algo.fit(X_train,y_train)
    algo.predict(X_test)
    score=algo.score(X_test,y_test)
    return score
```

```python
n_val=[2,8,15,25,35]
size=[0.3,0.4,0.5]
accur=[]
for i in size:
    accur=[]
    for j in n_val:
        a=n_estim(i,j)
        accur.append(a)
        
    #plt.figure(figsize=(5,10))
    plt.plot(n_val,accur,label=i)
    plt.xlabel('n_estimator values',fontsize=14)
    plt.ylabel('score',fontsize=14)
    plt.title('Test data set n_estimators=[2,8,15,25,35]',fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend()
    plt.plot()
```
